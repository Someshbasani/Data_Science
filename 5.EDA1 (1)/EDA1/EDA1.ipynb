# Outliers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numerical_cols = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width']

# Function to detect and treat outliers using IQR
def treat_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df

# Visualize outliers before treatment
plt.figure(figsize=(20, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(x=df[col])
    plt.title(f'Before Treatment: {col}')
plt.tight_layout()
plt.show()

# 
```

```python
#Treat outliers for all numerical columns
for col in numerical_cols:
    df = treat_outliers_iqr(df, col)

# Visualize outliers after treatment
plt.figure(figsize=(20, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(x=df[col])
    plt.title(f'After Treatment: {col}')
plt.tight_layout()
plt.show()
```

## 4. Pattern Recognition and insights
### Identify any correlations between variables and discuss their potential implications.

```python

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Interpretation
### 1.Strong Positive Correlation: Values close to 1 indicate a strong positive relationship.

### 2.Strong Negative Correlation: Values close to -1 indicate a strong negative relationship.

### 3.Weak or No Correlation: Values close to 0 indicate little to no relationship.

# EX:

### 1. LB and AC: If there is a positive correlation, it could suggest that higher baseline fetal heart rates are associated with more accelerations, which may indicate fetal well-being.

### 2. UC and DL: A negative correlation might imply that stronger uterine contractions are associated with fewer late decelerations, which could be a sign of fetal distress.

```python
# 4.2 Trends and Patterns
# Example: Plotting LB over time (assuming rows are ordered chronologically)
plt.figure(figsize=(10, 6))
plt.plot(df['LB'], color='blue', label='Baseline Fetal Heart Rate (LB)')
plt.title('Trend of Baseline Fetal Heart Rate (LB) Over Time')
plt.xlabel('Time')
plt.ylabel('LB')
plt.legend()
plt.show()
```
