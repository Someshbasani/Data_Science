### Provide a brief interpretation of these statistics.

```python
      Volume:
      The average (mean) volume sold is about 5 units, with most transactions selling between 1 and 6 units (the 25th percentile is 3 and the median is 4).
    However, the maximum volume is 31, which suggests a few transactions involve much higher volumes.This indicates a slightly right‐skewed distribution.

    Avg Price: 
      Although the median price is 1,450, the mean is much higher (around 10,453), and the maximum is 60,100. This large difference between the mean and
    the median, along with a high standard deviation, suggests that while most prices are relatively low, a few very expensive items are inflating the
    average.

    Total Sales Value:  
      The mean total sales value is about 33,813 with a median of 5,700. Again, the higher mean compared to the median and a large standard deviation 
    indicate a right-skewed distribution—most transactions have lower sales values, but a small number have very high sales values (up to 196,400).

    Discount Rate (%):  
      The discount rates are more consistent, ranging from about 5% to nearly 20% (with a median around 16.6% and mean around 15.2%). The spread here is
    much narrower, indicating that most transactions receive similar discount percentages.

      Discount Amount and Net Sales Value:  
      Both of these metrics show similar patterns to Total Sales Value. The mean discount amount (about 3,346) is much higher than the median
    (around 989), and the mean net sales value (about 30,466) is significantly higher than the median (around 4,678). This again points to a distribution 
    where most transactions have lower values, but a few very high-value transactions drive up the averages.

```

# Data Visualization

## Histograms for numerical columns

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numerical_columns = ['Volume','Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
df[numerical_columns].hist(bins=20, figsize=(15,10))
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

# Boxplots for Numerical Variables

```python
plt.figure(figsize=(15, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[column])
    plt.title(column)
plt.suptitle('Boxplots of Numerical Columns')
plt.tight_layout()
plt.show()
```

#  Bar Chart Analysis for Categorical Columns

```python
categorical_columns = ['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']
plt.figure(figsize=(20, 15))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(3, 2, i)
    df[column].value_counts().plot(kind='bar')
    plt.title(column)
    plt.xticks(rotation=45)
plt.suptitle('Bar Charts of Categorical Columns')
plt.tight_layout()
plt.show()
```

# Standardization of Numerical Variables

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df[numerical_columns].describe()
```

```python
import matplotlib.pyplot as plt
df[numerical_columns].hist(bins=20, figsize=(15,10))
plt.suptitle('Distributions After Standardization')
plt.show()
```

# Conversion of Categorical Data into Dummy Variables

```python
df_encoded = pd.get_dummies(df, columns= categorical_columns, drop_first =True)
```
