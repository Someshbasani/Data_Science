```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data = pd.read_excel('Onlineretail.xlsx', header=None)
```

```python
df1 = data.fillna('')
```

```python
df = df1.drop_duplicates()
```

```python
transactions = df[0].apply(lambda x: x.split(',')).tolist()
```

```python

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_) #df_encoded = pd.DataFrame(te_ary.astype(int), columns=te.columns_)
```

```python
df_encoded
```

```python
df_encoded.to_csv('preprocessed_transactions.csv', index=False)
```

```python
Data = pd.read_csv('preprocessed_transactions.csv')
```

```python
Data.head()
```

```python
df = pd.get_dummies(Data, dtype = 'int')
df.head()
```

```python
df.dtypes
```

```python
# Apriori Algorithm
```

```python
Frequent_itemsets = apriori(df, min_support = 0.01, use_colnames = True)
Frequent_itemsets
```

```python
rules = association_rules(Frequent_itemsets, metric = 'lift', min_threshold = 1.0)
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
df = rules[rules.lift>2]
df
```

```python
df.sort_values('confidence',ascending = False)[0:20]
```

```python
3. What are Some Limitations or Challenges of Association Rule Mining?
Too Many Rules

The algorithm can generate thousands of rules, making it hard to interpret results.
Solution: Use thresholds like minimum support, confidence, and lift.
Choosing the Right Thresholds

If support is too high, we miss interesting rare associations.
If support is too low, we get too many irrelevant rules.
Handling Large Datasets

Processing large datasets is computationally expensive.
Solution: Use optimized algorithms like Apriori or FP-Growth.
Redundant or Useless Rules

Many rules are trivial (e.g., "Milk → Bread" is expected).
Solution: Use lift or interestingness measures to filter rules.
Correlation vs. Causation

Association rules show correlation, not causation.
Example: "Ice Cream → Sunglasses" does not mean buying ice cream causes sunglasses purchases; they occur together due to summer.
```
